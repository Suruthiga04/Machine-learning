```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("abalone.csv")
df.head()
```

```python
data=pd.DataFrame()
data["x"]=df["LongestShell"]
data["y"]=df["Diameter"]
data.head()
```

```python
import math
import numpy as np

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def distance(d1,d2):
    return math.sqrt(math.pow(float(d1),2)+ math.pow(float(d2),2))

m = 100
P = [Point(0, 0) for _ in range(m)]

for i in range(m):
    P[i].x, P[i].y = data.loc[i, 'x'], data.loc[i, 'y']

k = int(input("Enter the number of clusters: "))

px=[]
py=[]
for i in range(m):
    px.append(P[i].x)
    py.append(P[i].y)

rtemp=[]
pcost=float('inf')
res=[]
x=100
fcost=0
while x>0:
    rp_indices = np.random.choice(m, size=k, replace=False).tolist()  # Randomly select k indices from the dataset
    while rp_indices in rtemp:
      rp_indices = np.random.choice(m, size=k, replace=False).tolist()
    rtemp.append(rp_indices)
    rp = [Point(0, 0) for _ in range(k)]
    for i in range(k):
      rp[i].x, rp[i].y = data.loc[rp_indices[i], 'x'], data.loc[rp_indices[i], 'y']
    rx=[]
    ry=[]
    for i in range(k):
      rx.append(rp[i].x)
      ry.append(rp[i].y)
    arr = [[] for _ in range(k)]
    mp = Point(0,0)
    count = [0] * k
    cost=0
    for i in range(m):
        d=[]
        for j in range(k):
            d.append(distance(P[i].x-rp[j].x,P[i].y-rp[j].y))
        c = d.index(min(d))
        cost+=min(d)
        arr[c].append(P[i])
        count[c] += 1

    if cost==pcost:
      print("Success")
      print(res)
      break
    elif cost<pcost:
      print("Cost lesser...")
      pcost=cost
      fcost=cost
      res=arr
    else:
      print("Cost greater...Leave this and go to next")
    x-=1
    print(x)
print("Cost: ",fcost)
s=1
for i in res:
  print("Cluster: ",s)
  for j in i:
    print(j.x," ",j.y)
  s+=1
```
